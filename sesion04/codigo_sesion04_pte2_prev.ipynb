```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport
import sweetviz as sv
import seaborn as sns
```

```python
# La raw URL del CSV en GitHub
url = 'https://raw.githubusercontent.com/EduardoSelimMM/20260101_AnalisisExploratorioDatos/refs/heads/main/sesion02/credit_risk_dataset.csv'

# Se lee el CSV y se guarda en un pandas DataFrame
df = pd.read_csv(url)
```

```python
df_casa_propia = df[df["person_home_ownership"] == "OWN"].copy()
df_no_casa_propia = df[df["person_home_ownership"] != "OWN"].copy()
```

```python
casa_propia_profile = ProfileReport(df_casa_propia, title = "Comportamiento de clientes con casa propia")
no_casa_propia_profile = ProfileReport(df_no_casa_propia, title = "Comportamiento de clientes sin casa propia")
```

```python
comparo_reportes = casa_propia_profile.compare(no_casa_propia_profile)
```
